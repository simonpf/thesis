```python
from h5py import File
from matplotlib.colors import LogNorm
import cartopy.crs as ccrs
from matplotlib.cm import ScalarMappable


norm = LogNorm(1e-1, 1e1)

class ImergAnimation:
    def __init__(self, ax, files):
        self.files = files
        ax.set_xlim([-180, 180])
        ax.set_ylim([-90, 90])
        ax.coastlines(color="w")
        ax.set_aspect(1.0)
        
        ax.set_xticks(np.linspace(-180, 180, 10))
        ax.set_xlabel("Longitude [$\si{\degree}$ E]")
        ax.set_yticks(np.linspace(-90, 90, 6))
        ax.set_ylabel("Latitude [$\si{\degree}$ N]")
        

    def __call__(self, i):
        
        print(f"{i} / {len(self.files)}")
        # Load precip data
        filename = self.files[i]
        if filename.suffix.startswith(".nc"):
            data = xr.load_dataset(filename)
            precip = np.maximum(np.nan_to_num(data.precipitationCal[0, :].T, 0.0), 1e-2)
            lons = data.lon.data
            lats = data.lat.data
            
        else:
            with File(self.files[i], "r") as data:
                precip = np.maximum(data["Grid/precipitation"][:].T, 1e-2)
                lons = data["Grid/lon"][:]
                lats = data["Grid/lat"][:]
            print(precip.min(), precip.max())
                
        lon_min, lon_max = lons.min(), lons.max()
        lat_min, lat_max = lats.min(), lats.max()
        m = ax.imshow(precip, origin="lower", extent=[lon_min, lon_max, lat_min, lat_max], cmap="rocket", norm=norm)
        ax.coastlines(color="w")
        ax.set_aspect(1.0)
            
        filename = self.files[i].name
        date = filename.split(".")[4]
        year = date[:4]
        month = date[4:6]
        day = date[6:8]
        hour = date[10:12]
        minute = date[12:14]
        date = np.datetime64(f"{year}-{month}-{day}T{hour}:{minute}:00")
        ax.set_title(date)
                            
        return m,
```

```python
from matplotlib.gridspec import GridSpec

f = plt.figure(figsize=(13.5, 6))
gs = GridSpec(1, 2, width_ratios=[1.0, 0.05])

ax = f.add_subplot(gs[0, 0], projection=ccrs.PlateCarree())
animator = ImergAnimation(ax, imerg_files)

m = ScalarMappable(cmap="rocket", norm=norm)
cax = f.add_subplot(gs[0, 1])
plt.colorbar(m, cax=cax, label="Mean precipitation rate [$\si{\milli \meter \per \hour}$]")
```
